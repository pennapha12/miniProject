```python
# รีเซ็ต index เพื่อให้แต่ละชื่อคอลัมน์เป็นแถวใน DataFrame ใหม่
df_reset = df.melt(var_name='Table', value_name='Column')

# นับจำนวนการปรากฏตัวของแต่ละคอลัมน์
column_counts = df_reset['Column'].value_counts()

# หาคอลัมน์ที่ปรากฏตัวมากกว่าหนึ่งครั้ง
possible_fks = column_counts[column_counts > 1]

# สร้าง DataFrame สำหรับคอลัมน์ที่เป็นไปได้ว่าเป็น Foreign Key
fk_rows = []

# เพิ่มข้อมูลลงในรายการแถว
for column in possible_fks.index:
    tables = df_reset[df_reset['Column'] == column]['Table'].tolist()
    fk_rows.append({'Column': column, 'Tables': ', '.join(tables)})

# สร้าง DataFrame จากรายการแถว
fk_df = pd.DataFrame(fk_rows)

# แสดงผลลัพธ์
print("Possible Foreign Key columns and their tables:")
print(fk_df)

# ถ้าต้องการจัดรูปแบบเพิ่มเติม (เช่น การแสดงข้อมูลแบบแยกคอลัมน์)
for index, row in fk_df.iterrows():
    print(f"\nColumn '{row['Column']}' appears in the following tables:")
    tables = row['Tables'].split(', ')
    for table in tables:
        print(f"- {table}")
```

```python
import polars as pl

# Define the conversion function
def convert_str_to_datetime(df, column_name, datetime_format='%Y-%m-%d %H:%M:%S'):
    return df.with_columns(
        pl.col(column_name).str.strptime(pl.Datetime, format=datetime_format).alias('datetime')
    )

# Apply the pipeline
inv = (
    invoices.pipe(rename_col, {"InvoiceId": "Inv_Id"})
    .pipe(rename_col, {"CustomerId": "Cus_Id"})
    .pipe(rename_col, {"InvoiceDate": "Inv_D"})
    .pipe(rename_col, {"BillingAddress": "B_Ad"})
    .pipe(rename_col, {"BillingCity": "B_Ci"})
    .pipe(rename_col, {"BillingState": "B_St"})
    .pipe(rename_col, {"BillingCountry": "B_Coun"})
    .pipe(rename_col, {"BillingPostalCode": "B_Pos"})
    .pipe(sort, 'Total')
    .pipe(convert_str_to_datetime, 'Inv_D')
)

# Inspect the final DataFrame
print(inv)
```

```python
import polars as pl

# Assuming rename_col and convert_str_to_datetime are already defined as per previous discussions

def split_month_year(df, datetime_col):
    # Ensure the datetime_col is in datetime format
    df = df.with_columns(
        pl.col(datetime_col).cast(pl.Datetime).alias(datetime_col)
    )
    
    # Add new columns for month and year
    return df.with_columns([
        pl.col(datetime_col).dt.month().alias(f"{datetime_col}_Month"),
        pl.col(datetime_col).dt.year().alias(f"{datetime_col}_Year")
    ])

# Your pipeline
inv = (
    invoices.pipe(rename_col, {"InvoiceId": "Inv_Id"})
    .pipe(rename_col, {"CustomerId": "Cus_Id"})
    .pipe(rename_col, {"InvoiceDate": "Inv_D"})
    .pipe(rename_col, {"BillingAddress": "B_Ad"})
    .pipe(rename_col, {"BillingCity": "B_Ci"})
    .pipe(rename_col, {"BillingState": "B_St"})
    .pipe(rename_col, {"BillingCountry": "B_Coun"})
    .pipe(rename_col, {"BillingPostalCode": "B_Pos"})
    .pipe(sort, 'Total')
    .pipe(convert_str_to_datetime, 'Inv_D')  # Convert to datetime format
    .pipe(split_month_year, 'Inv_D')          # Use the correct column name
)

print(inv.head())
```

```python

# เลือกคอลัมน์ที่จำเป็นจากแต่ละ DataFrame
Ge_sel = Ge_f.select(['ge_id', 'ge_name'])
Al_sel = Al_f.select(['al_id', 'al_ti', 'ar_id'])
Ar_sel = Ar_f.select(['ar_id', 'ar_name'])
Invt_sel = Invt_f.select(['inv_id', 'tr_id'])  # ปรับชื่อคอลัมน์ตามที่ตรวจสอบ
Inv_sel = Inv_f.select(['inv_id', 'total','date_time_month','date_time_year'])
Tr_sel = Tr_f.select(['tr_id', 'tr_name', 'ge_id', 'al_id'])

df = (Al_sel.join(Tr_sel, on='al_id', how='inner') \
           .join(Ge_sel, on='ge_id', how='inner') \
           .join(Ar_sel, on='ar_id', how='inner') \
           .join(Invt_sel, on='tr_id', how='inner') \
           .join(Inv_sel, on='inv_id', how='inner') \
           .sort(by='total', descending=True) \
           .group_by('date_time_year','date_time_month')  # แบ่งกลุ่มตามปี \
           .agg([ \
               pl.col('total').sum().alias('total_sum'),  # ผลรวมของ total ในแต่ละกลุ่ม \
               pl.count().alias('count'),  # จำนวนรายการในแต่ละกลุ่ม \
               pl.col('al_ti').first().alias('album_name'),  # แสดงชื่ออัลบั้มแรกในแต่ละกลุ่ม \
               pl.col('tr_name').first().alias('first_track_name'), # แสดงชื่อเพลงแรกในแต่ละกลุ่ม \
               pl.col('ar_name').first().alias('artists_name')
           ]) \
           .sort(by=['date_time_year','date_time_month'])  # เรียงลำดับตามปี
)
print(df)
```
